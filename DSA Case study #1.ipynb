# 4. The top 5 cars based on their sales price.


```python
top5=data.nlargest(5,['Resell Price'])
print(top5['Make'])
```

# 5. The least 5 cars based on their Resell price.

```python
least5=data.nsmallest(5,['Resell Price'])
print(least5['Make'])
```
